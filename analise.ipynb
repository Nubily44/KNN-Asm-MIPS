```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_raw = pd.read_csv('Xtrain.csv', header = None)
df_raw.head()
```

```python
df_raw2 = pd.read_csv('Xtest.csv', header = None)
df_raw2.head()
```

```python
df = df_raw.copy()
df.columns = ['value']
df.insert(0, 'Index', df.index)
df.head()
```

```python
df2 = df_raw2.copy()
df2.columns = ['value']
df2.insert(0, 'Index', df.index)
df2.head()
```

```python
plt.figure(figsize=(20, 10))
palette = sns.color_palette("Paired", 4)

sns.lineplot(data=df, x=df.index, y = df['value'], marker='o', color = palette[0], label = 'Treino')
#sns.lineplot(data=df, x=df.index, y = df['value'] * 10, marker='o', color = palette[1], label = 'Treino x 10')
#sns.lineplot(data=df2, x=df2.index, y = df2['value'], marker='o', color = palette[2], label = 'Teste')
sns.lineplot(data=df2, x=df2.index, y=np.log(df2['value']), marker='o', color=palette[2], label='Log Teste')
plt.xlabel('Tempo')
plt.ylabel('Bolsa')
plt.title('CLODOALDOOOO')
plt.show()
```
